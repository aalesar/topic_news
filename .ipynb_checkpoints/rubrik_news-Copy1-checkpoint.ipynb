```python
FileList = glob.glob('*.csv') 
print(FileList) 
end_work=[]
err_log=[]
proverka=[]
propusk=[]
new_data=[]
```

```python
len(FileList)
```

```python
len(end_work)
```

```python
Work_list=list(set(FileList)-set(end_work))
kom_dict={}
for fname in Work_list: 
   
    df = pd.read_csv(fname) 
    df[['date','URL']]=df[['URL','date']]
    
    print(fname)
    print(df['URL'].map(lambda x: x.split('/')).head())
    
    z= input()
    if z=='':
        print('Введите число')
        z=input()
    kom=int(z)
    if kom==0:
        end_work.append(fname)
        propusk.append(fname)
        print(fname)
        continue
    elif kom==99:break
    elif kom==23:
        proverka.append(fname)
        continue    
    else:
       
        try:
            df['topics'] = df['URL'].map(lambda x: x.split('/')[kom])
            print(fname)
            print(df.topics.head())
            df.to_csv('OUT/'+fname, encoding='utf-8',index=False)
            end_work.append(fname)
            new_data.append(fname)
            kom_dict[fname]=kom
            
            print('++++++++++++++++++OK++++++++++++++++++++')
        except:
            
            err_log.append(fname)
            print(err_log)
            print('-------------------ОШИБКА---------------')
           
            end_work.append(fname)
            
            
with open('OUT/result.txt','w') as result:
    result.write('Всего просмотрено файлов: '+str(len(end_work))+'\n\n')
    result.write('Данные изменены в : '+str(len(new_data))+' файлах'+'\n\n')
    result.write('Разная структура URL в: '+str(err_log)+'\n\n')
    result.write('Проверить: '+str(proverka))+'\n\n'
    result.write('Всего просмотрено файлов: '+str(end_work)+'\n\n')
    result.write('Всего просмотрено файлов: '+str(kom_dict)+'\n\n')
    

print('Обработано',len(end_work))
print('Ошибок',len(err_log))
print('Проверить',len(proverka))
```
